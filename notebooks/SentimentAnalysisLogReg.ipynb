```python
import pickle

# carregar o modelo treinado
with open('model/logistic_regression_model.pkl', 'rb') as f:
    clf = pickle.load(f)

# carregar o CountVectorizer
with open('model/count_vectorizer.pkl', 'rb') as f:
    vect = pickle.load(f)

# pré-processar a frase de entrada
def preprocess_text(text):
    text_vect = vect.transform([text])
    return text_vect

# prever o sentimento da frase
def predict_sentiment(text):
    text_vect = preprocess_text(text)
    sentiment = clf.predict(text_vect)
    if sentiment == 1:
        return 'Positive'
    elif sentiment == 0:
        return 'Neutral'
    else:
        return 'Negative'
```

```python
import pandas as pd

df = pd.read_csv('../Files/data_to_analyze.csv')

df
```

```python
import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

# função de remover stopwords do texto
def remove_stopwords(text):
    if isinstance(text, str):
        words = text.split()
        filtered_words = [word for word in words if word.lower() not in stop_words]
        return ' '.join(filtered_words)
    else:
        return ''
    
sentiments = []
for index, row in df.iterrows():
    text = remove_stopwords(row['Total_Review'])
    sentiment = predict_sentiment(text)
    sentiments.append(sentiment)

# add coluna sentiment
df['sentiment'] = sentiments

# uso
# input_text = "i hate the room"
# predicted_sentiment = predict_sentiment(input_text)
# print("Sentimento previsto:", predicted_sentiment)
```
