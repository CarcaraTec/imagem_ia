```python
data = pd.read_csv(r'..\Files\Hotel_Reviews.csv', low_memory=False,delimiter=',', encoding='iso-8859-1', decimal='.')
data.info()
```

```python
# cria novo dataframe a partir das colunas que eu quero
df = data[['Review_Date','Hotel_Name',
           'Negative_Review', 'Positive_Review', 'Reviewer_Score',
           'Tags', 'lat', 'lng']]
df.sample(10)
```

```python
# transforma coluna Review_Date para datetime
df.loc[:, 'Review_Date'] = pd.to_datetime(df['Review_Date'], format='%m/%d/%Y')
df.info()
```

```python
# substitui as avaliações "No negative" e "No positive" por string vazia
df.loc[:, 'Positive_Review'] = data.Positive_Review.apply(lambda x: x.replace('No Positive', ''))
df.loc[:, 'Negative_Review'] = data.Negative_Review.apply(lambda x: x.replace('No Negative', ''))
df.sample(5)
```

```python
# une as colunas de avaliações negativas e positivas em uma só
df.loc[:,"Total_Review"] = df["Negative_Review"] + df["Positive_Review"]
df.sample(5)
```

```python
# classifica avaliações com nota inferior a 6 como ruins e maiores como
df.loc[:,"review_type"] = df["Reviewer_Score"].apply(
    lambda x: "Bad" if x < 4 else ("Neutral" if x < 7 else "Good")
)

df
```

```python
# cria um dataframe somente com o comentário e sua classificação
df_reviews = df[["Total_Review", "review_type"]]
df_reviews
```

```python
# mostra os comentários positivos citados e a quantidade de cada um deles se repetidos
df_reviews[df_reviews.review_type == 'Good'].Total_Review.value_counts()
```

```python
# mostra os comentários negativos citados e a quantidade de cada um deles se repetidos
df_reviews[df_reviews.review_type == 'Bad'].Total_Review.value_counts()
```

```python
# separa as avaliações em good e bad
good_reviews = df[df.review_type == "Good"]
bad_reviews = df[df.review_type == "Bad"]
neutral_reviews = df[df.review_type == "Neutral"]
```

```python
# aleatoriza e seleciona aleatoriamente um número de avaliações positivas 
# igual ao número de avaliações negativas e neutras
good_df = good_reviews.sample(n=len(bad_reviews) + len(neutral_reviews), random_state=48)

# Concatena os DataFrames de avaliações positivas e neutras com as avaliações negativas
df_review_resampled = pd.concat([good_df, neutral_reviews, bad_reviews], ignore_index=True)

# Verifica a forma do DataFrame resultante
df_review_resampled
```
